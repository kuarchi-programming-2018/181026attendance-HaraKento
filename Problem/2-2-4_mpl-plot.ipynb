## matplotlibを使ってグラフをプロット

まずはデータを読み込み，それをX, Yのデータにして返す関数を作りましょう。
csv とはcomma separated valuesです。テキストで開いてみればわかります。エクセルでも開けます。

```python
import matplotlib.pyplot as plt
% matplotlib inline
import numpy as np
x = np.arange(-3,3,0.1)
y = np.sin(x)
plt.title("title")
plt.xlabel("X-LABEL")
plt.ylabel("Y-LABEL")
plt.plot(x,y)
plt.show()
```

```python
import csv
def read_data(file_name):
    reader = csv.reader(open(file_name,'r'))
    X=[]
    Y=[]
    for row in reader:
        X.append(float(row[0]))
        Y.append(float(row[1]))
    return X,Y    
file_name = 'data3.csv'  
x,y=read_data(file_name)
print(x)
print(y)

def draw_graph(X, Y, title, xlabel, ylabel):
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.plot(X,Y)
    plt.show()  # グラフの描画
title="a"
xl="x"
yl="y"
X,Y=x,y
# 関数read_dataによりデータを取得
# タイトル，ラベル
draw_graph(X, Y, title, xl, yl) # 関数draw_graphによりグラフ描画
```
